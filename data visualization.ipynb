```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np

df = pd.read_csv("train.csv")

for column in df.columns: #replace NaN by random data
    df[column] = df[column].fillna(np.random.choice(df[column].unique()))
    
cols = df.columns[-3:]
fig, axs = plt.subplots(1, 3, figsize=(20,  10))
    
for i in range(3): #build barplot
    d = dict(df[cols[i]].value_counts())
    axs[i].set_title(f"{cols[i]}")
    axs[i].bar(d.keys(), d.values())
plt.show()
```
