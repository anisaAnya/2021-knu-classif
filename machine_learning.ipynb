# 2021-Jan-29 (Fri) – Random Forest

### Removing outliers from our dataframe

```python
from scipy import stats

ml_df = main_df.copy()
num_df = ml_df[ml_df.columns[:-3]]

num_df = num_df[(np.abs(stats.zscore(num_df)) < 3).all(axis=1)]
indexes = list(num_df.index)

ml_df = ml_df.loc[indexes]
ml_df[ml_df.columns[:-3]] = num_df


fl_df = normalized_df.loc[indexes]

ml_df



```

### Split daraframe on train and test datasets

```python
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

X = ml_df[ml_df.columns[:-1]]
Y = ml_df['customer_category']

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0)

randforest = RandomForestClassifier(n_estimators=200)
randforest.fit(X_train, y_train)

rf_pred = randforest.predict(X_test)

print("Accuracy:",metrics.accuracy_score(y_test, rf_pred))
print("Precision:",metrics.precision_score(y_test, rf_pred))
print("Recall:",metrics.recall_score(y_test, rf_pred))
```

# 2021-Feb-2-4 (Tue-Thu) – Model Evaluation, ROC Curve

### Main calculations (roc curve, roc AUC score, cross-validation)

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

randforest = RandomForestClassifier()
nearneighbor = KNeighborsClassifier(n_neighbors=6)
naivebayes = GaussianNB()
supvector = SVC(kernel="linear", probability=True)
logregression = LogisticRegression()

models = [
    ["Random Forest ", randforest],
    ["K neighbors ", nearneighbor],
    ["Naive Bayes ", naivebayes], 
    ["Support vector machine ", supvector],
    ["Regression ", logregression]
]

pred_values = []
proba_values = []
roc_auc_score_values = []
roc_curve_values = []
cross_val_scores = []

for model in models:
    print(models.index(model) * 100 // len(models), "% ...")
    model[1].fit(X_train, y_train)
    
    pred = model[1].predict(X_test)
    proba = model[1].predict_proba(X_test)[:, 1]
    auc = metrics.roc_auc_score(y_test, proba)
    pr, tpr, thresholds = metrics.roc_curve(y_test, proba)
    
    cross_score = cross_val_score(model[1], X, Y, cv=10, scoring="accuracy").mean()
    
    pred_values.append([model[0], pred])
    proba_values.append([model[0], proba])
    roc_auc_score_values.append([model[0], auc])
    roc_curve_values.append([model[0], [pr, tpr]])
    cross_val_scores.append([model[0], cross_score])
    
print("Training done")
```

### Print roc "area under curve" score, cross-validation accuracies and plot roc curve

```python

print("Accuracy for each algorithm:\n")
for value in pred_values:
    print(value[0], metrics.accuracy_score(y_test, value[1]))

print("\n")
    
print("Area under each curve after each algorithm:\n")
for value in roc_auc_score_values:
    print(*value)
    
print("\n")
    
print("Cross validation accuracy for each algorithm:\n")
for value in cross_val_scores:
    print(*value)
    
for value in roc_curve_values:
    plt.plot(value[1][0], value[1][1], marker=".", label=value[0])
    
plt.xlabel("False Positive")
plt.ylabel("True Positive")
plt.legend()
plt.show()

```

```python
def send_model_to_user():
    pass # because we dont really have users

def get_model_back():
    pass # because we don't really have users

clf = RandomForestClassifier(warm_start=True)

for i in range(len(ml_df)):
    clf.n_estimators += 1
    
    send_model_to_user() # send model
    
    row = ml_df.iloc[i]

    clf.fit([row[:-1]], [row[-1]]) # this row creates a new tree for a user (this tree will be added in forest)
    
    get_model_back() # get model back

pred = clf.predict(X_test)
metrics.accuracy_score(y_test, pred)
```
