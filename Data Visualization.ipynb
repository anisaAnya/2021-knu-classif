DATA VISUALIZATION

```python
for col in df.columns[1:8]:
    df[col] = df[col].fillna(df[col].mean())
for col in df.columns[9:12]:
    df[col] = df[col].fillna(np.random.choice(df[col].unique()))  
```

```python
fig, axs = plt.subplots(1, 3, figsize=(25,  5))
for i in range (3):
    dct = dict(df[df.columns[9:12][i]].value_counts())
    axs[i].set_title(f"{df.columns[i+9]}")
    axs[i].bar(dct.keys(), dct.values())
    


```
