```python
'''Энкодинг'''

LE = LabelEncoder()
sample_data[columns] = sample_data[columns].apply(lambda columns: LE.fit_transform(columns))
sample_data
```

```python
'''Забыл посчитать mean, standard deviation, correlation matrix и frequency'''
columns = list(sample_data.columns)
for i in range(0,len(columns)):
    try:
        var = sample_data[columns[i]]
        ans = statistics.mean(var)
        print('Mean of {0} = {1}'.format(columns[i],ans))
    except TypeError:
        continue
print("\n")
for i in range(0,len(columns)):
    try:
        mean = sample_data[columns[i]].value_counts()
        keys = list(mean.keys())
        print(keys[0], mean[keys[0]])
    except TypeError:
        continue
```

```python
columns = list(sample_data.columns)
for k in range(0,len(columns)):
    try:
        nu = sample_data[columns[k]]
        summ_nu = 0
        summ_sigma = 0
        for i in range(len(nu)):
            summ_nu += nu[i]
        nu_res = summ_nu/len(nu)
        for i in range(len(nu)):
            summ_sigma += ((nu[i] - nu_res)*(nu[i] - nu_res))
        sigma_res = math.sqrt(summ_sigma/len(nu))
        print(columns[k], sigma_res)
    except TypeError:
        continue
```

```python
columns = list(sample_data.columns)
for k in range(0,len(columns)):
    try:
        freq = sample_data[columns[k]].value_counts()
        print(freq)
    except TypeError:
        continue
```
