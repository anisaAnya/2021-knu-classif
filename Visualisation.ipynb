```python
'''Создаю датасет без обьектных значений, для того чтоб нормализировать данные'''

df_data=pd.DataFrame([sample_data.customer_visit_score, sample_data.customer_product_search_score,sample_data.customer_ctr_score,sample_data.customer_stay_score,sample_data.customer_frequency_score,sample_data.customer_product_variation_score,sample_data.customer_order_score,sample_data.customer_affinity_score])
columns=['customer_visit_score', 'customer_product_search_score','sample_data.customer_ctr_score','sample_data.customer_stay_score','sample_data.customer_frequency_score','sample_data.customer_product_variation_score','sample_data.customer_order_score','sample_data.customer_affinity_score']
df_data
```

```python
'''Нормализация'''

def min_max_scaling(df_data):
    df_norm = df_data.copy()
    for column in df_data:
        df_norm[column] = (df_norm[column] - df_norm[column].min()) / (df_norm[column].max() - df_norm[column].min())
        
    return df_norm
    
df_normalized = min_max_scaling(df_data)

df_normalized
```

```python
sns.heatmap(df_normalized)
```
